```python
#weightwall_all.shape #[i for i in weightwall_all[:,740] if i > 0]
weightlandmark_all.shape
[i for i in weightlandmark_all[:,740] if i > 0]
bestlandmark = np.argmax(weightlandmark_all[:,740])
bestlandmark = np.array(bestlandmark)
```

```python
#function to transfer position index into 2D coordinates
def get_coordinates(position_index_array):
    x = [position%N_axis for position in position_index_array]
    y = [(position - (position%N_axis))/N_axis for position in position_index_array]
    return x,y
```

```python
#getting 2D illustrates for walls
wall_index = np.where(weightwall_all[:,740]>0)
bestlandmark = [np.argmax(weightlandmark_all[:,740])]
landmark_index = np.array(bestlandmark)
# obtaining wall and landmarkcoordinates 
x_wall, y_wall = get_coordinates(wall_index)
x_landmark,y_landmark = get_coordinates(landmark_index)

########################visualizing learned wall and landmark#######################
#########blue is the estimated wall, red is the estimated landmark###############
plt.rcParams['figure.figsize'] = [10, 7]
plt.scatter(x_wall, y_wall, marker='o')
plt.scatter(x_landmark,y_landmark, color = "red", marker='o')
```

```python
# obtaining IMU collected collision index 
x_colli, y_colli = get_coordinates(collision_index)
########################visualizing collision positions gained from IMU #######################
plt.scatter(x_colli, y_colli, marker='o')
```

```python
#############################plot the movement path based on IMU returned calibration ############
#get coordinates of IMU returned position index
trace = np.array(r)
x_trace, y_trace = get_coordinates(trace)
plt.plot(x_trace, y_trace, 'red', lw=1)
plt.scatter(x_trace, y_trace, s=5)
```

```python
'''comparing firing time of collision detection neuron(green), wall detection neuron(blue), 
landmark detection neuron(red) and non-collision neuron (purple)'''
#### theoretically, collision detection neuron's activity should be the sum of wall and landmark detectio neurons
##additionally noncollision neuron should be firing when the other 3 neurons are not

plt.rcParams['figure.figsize'] = [22,5]
#plot wall detection
fig = plt.figure()
plt.scatter(spikemon_wall_t,np.zeros(len(spikemon_wall_t)),color = "blue",s =5)
plt.scatter(spikemon_landmark_t,np.repeat(0.5,len(spikemon_landmark_t)),color = "red",s =5)
plt.scatter(spikemon_CD_t,np.repeat(-0.5,len(spikemon_CD_t)),color = "green",s=5)
plt.scatter(spikemon_noncol_t,np.repeat(-1,len(spikemon_noncol_t)),color = "purple",s=5)

pylab.xlim([4600,5000])
pylab.ylim([-1.2,1])

plt.tick_params(axis='x', labelsize=15)
plt.tick_params(axis='y', labelsize=15)
plt.xlabel('Time (s)' ,fontsize=15)
plt.ylabel('3 types of detection Neurons  ', fontsize=15)
```

```python
'''2. Heading direction neuron'''
#plot heading direction
fig = plt.figure()
plt.plot(spikemon_HD_t,spikemon_HD_i, '.k')
fig.set_size_inches(9.5, 7.5)
pylab.ylim([0,N_HD])
plt.tick_params(axis='x', labelsize=20)
plt.tick_params(axis='y', labelsize=20)
plt.xlabel('Time (s)' ,fontsize=20)
plt.ylabel('Neuron index ', fontsize=20)
```
